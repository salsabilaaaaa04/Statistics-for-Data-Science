# Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)
significant level = 0.05

H0: There is no significant difference in median value of houses bounded by the Charles river.
M1 = M0 , where M1 is median value of house bounded by the charles river and M0 is median value of house not bounded by the charles river

Ha: There is a significant difference in median value of houses bounded by the Charles river.
M1 != M0

```python
scipy.stats.levene(boston_df[boston_df['CHAS'] == 1]['MEDV'], boston_df[boston_df['CHAS']==0]['MEDV'], center = 'mean')
```

given p-value is less than 0.05, we can assume variance not equal

```python
scipy.stats.ttest_ind(boston_df[boston_df['CHAS'] == 1]['MEDV'], boston_df[boston_df['CHAS']==0]['MEDV'], equal_var = False)
```

given p-value less than 0.05, reject null hypothesis and conclude that there is a significant difference in the median values of houses bounded by the Charles River.

# Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)


AGE will be categorized into 3 groups :
35 and less
between 35 and 75
75 and more

H0: There is no difference in median values of houses for each proportion of owner occupied units built prior to 1940
Ha: There is a difference in median values of houses for each proportion of owner occupied units built prior to 1940.

```python
boston_df.loc[boston_df['AGE'] <= 35, 'age_group'] = '35 and less'
boston_df.loc[(boston_df['AGE'] > 35) & (boston_df['AGE'] < 75), 'age_group'] = 'between 35 and 75'
boston_df.loc[boston_df['AGE'] >= 75, 'age_group'] = '75 and more'
```

```python
scipy.stats.levene(boston_df[boston_df['age_group'] == '35 and less']['MEDV'], boston_df[boston_df['age_group'] == 'between 35 and 75']['MEDV'], boston_df[boston_df['age_group'] == '75 and more']['MEDV'], center='mean')
```

```python
lower = boston_df[boston_df['age_group'] == '35 and less']['MEDV']
middle = boston_df[boston_df['age_group'] == 'between 35 and 75']['MEDV']
upper = boston_df[boston_df['age_group'] == '75 and more']['MEDV']
```

```python
f_statistic, p_value = scipy.stats.f_oneway(lower, middle, upper)
print("F_Statistics: {0}, P-Value: {1}".format(f_statistic, p_value))
```

given p-value less than 0.05, reject H0 null hypothesis and conclude that there is a difference in median values of houses for each proportion of owner occupied units built prior to 1940.

# What is the impact of an additional weighted distance  to the five Boston employment centres on the median value of owner occupied homes? (Regression analysis)

H0: DIS has no impact on MEDV
Ha: DIS has an impact on MEDV



```python
# independent variable
x = boston_df['DIS']
# dependent variable
y = boston_df['MEDV']
# intercept beta
X = sm.add_constant(x)

model = sm.OLS(y,X).fit()
predictions = model.predict(X)

model.summary()
```

given the p value less than 0.05, we reject null hypothesis and conclude that there is evidence that there is impact on MEDV depending on the DIS. The impact is 1.0916, meaning that the as the DIS increases by one unit, the MEDV will increase approximately 1.0916 units.
